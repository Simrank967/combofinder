```python
import random as r
prodlist={'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35, 'p7':50,
               'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70,
               'p14':45}
upper=310
lower=290
result=set()
iterations=1000

for i in range(iterations):
  setsize=r.randint(2,len(prodlist)-1)
  combo=r.sample(list(prodlist.keys()),setsize)
  combo.sort()

  combosum=sum([prodlist[i] for i in combo])
  if combosum>=lower and combosum<=upper:
    result.add(tuple(combo))

for s in result:
  print(s,'\n')
```
